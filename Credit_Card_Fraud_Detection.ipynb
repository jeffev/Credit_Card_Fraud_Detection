## Prepare the dataset

Agora vamos dividir o conjunto de dados em conjuntos de treinamento e teste:

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Supondo que você já tenha os conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class', axis=1), df['Class'], test_size=0.3, random_state=42)

print("{} exemplos para treinamento, {} exemplos para teste.".format(
    len(X_train), len(X_test)))
```

```python
from sklearn.metrics import confusion_matrix

# Função para plotar a Matriz de Confusão
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', annot_kws={"size": 16})
    plt.title(f'Matriz de Confusão para {model_name}')
    plt.xlabel('Rótulos Previstos')
    plt.ylabel('Rótulos Verdadeiros')
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# Exemplo com Regressão Logística
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
predictions_logreg = logreg.predict(X_test)

plot_confusion_matrix(y_test, predictions_logreg, 'Regressão Logística')
```

```python
from sklearn.ensemble import RandomForestClassifier

# Exemplo com Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
predictions_rf = rf.predict(X_test)

plot_confusion_matrix(y_test, predictions_rf, 'Random Forest')
```

```python
from sklearn.tree import DecisionTreeClassifier

# Exemplo com Árvores de Decisão
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
predictions_dt = dt.predict(X_test)

plot_confusion_matrix(y_test, predictions_dt, 'Árvore de Decisão')
```

```python
from sklearn.svm import SVC

# Exemplo com SVM (Máquinas de Vetores de Suporte)
svm = SVC()
svm.fit(X_train, y_train)
predictions_svm = svm.predict(X_test)

plot_confusion_matrix(y_test, predictions_svm, 'SVM')
```

```python
from sklearn.neural_network import MLPClassifier

# Exemplo com Redes Neurais
nn = MLPClassifier(hidden_layer_sizes=(64, 64), max_iter=1000)  # Exemplo com duas camadas ocultas de 64 neurônios cada
nn.fit(X_train, y_train)
predictions_nn = nn.predict(X_test)

plot_confusion_matrix(y_test, predictions_nn, 'Redes Neurais')
```

```python
from sklearn.metrics import classification_report, roc_auc_score

# Avaliação
print("Regressão Logística:")
print(classification_report(y_test, predictions_logreg))
print("AUC-ROC Score:", roc_auc_score(y_test, predictions_logreg))

print("\nRandom Forest:")
print(classification_report(y_test, predictions_rf))
print("AUC-ROC Score:", roc_auc_score(y_test, predictions_rf))

print("Árvore de Decisão:")
print(classification_report(y_test, predictions_dt))
print("AUC-ROC Score:", roc_auc_score(y_test, predictions_dt))

print("\nSVM:")
print(classification_report(y_test, predictions_svm))
print("AUC-ROC Score:", roc_auc_score(y_test, predictions_svm))

print("\nRedes Neurais:")
print(classification_report(y_test, predictions_nn))
print("AUC-ROC Score:", roc_auc_score(y_test, predictions_nn))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Listas para armazenar as métricas de cada modelo
modelos = ['Regressão Logística', 'Random Forest', 'Árvore de Decisão', 'SVM', 'Redes Neurais']
precision = [precision_score(y_test, predictions_logreg), precision_score(y_test, predictions_rf),
             precision_score(y_test, predictions_dt), precision_score(y_test, predictions_svm),
             precision_score(y_test, predictions_nn)]
recall = [recall_score(y_test, predictions_logreg), recall_score(y_test, predictions_rf),
          recall_score(y_test, predictions_dt), recall_score(y_test, predictions_svm),
          recall_score(y_test, predictions_nn)]
f1 = [f1_score(y_test, predictions_logreg), f1_score(y_test, predictions_rf),
      f1_score(y_test, predictions_dt), f1_score(y_test, predictions_svm),
      f1_score(y_test, predictions_nn)]
roc_auc = [roc_auc_score(y_test, predictions_logreg), roc_auc_score(y_test, predictions_rf),
           roc_auc_score(y_test, predictions_dt), roc_auc_score(y_test, predictions_svm),
           roc_auc_score(y_test, predictions_nn)]

# Crie um DataFrame para facilitar a visualização
results_df = pd.DataFrame({
    'Modelo': modelos,
    'Precisão': precision,
    'Recall': recall,
    'F1-Score': f1,
    'AUC-ROC': roc_auc
})

# Plote os resultados usando um gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(data=results_df, x='Modelo', y='Precisão', color='skyblue', label='Precisão')
sns.barplot(data=results_df, x='Modelo', y='Recall', color='salmon', label='Recall')
sns.barplot(data=results_df, x='Modelo', y='F1-Score', color='green', label='F1-Score')
sns.barplot(data=results_df, x='Modelo', y='AUC-ROC', color='purple', label='AUC-ROC')
plt.title('Comparação de Métricas para Diferentes Modelos')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_curve

# Função para plotar a Curva ROC
def plot_roc_curve(y_true, y_scores, label):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc_score(y_true, y_scores):.2f})')

# Adicione as Curvas ROC para cada modelo
plot_roc_curve(y_test, logreg.predict_proba(X_test)[:, 1], 'Regressão Logística')
plot_roc_curve(y_test, rf.predict_proba(X_test)[:, 1], 'Random Forest')
plot_roc_curve(y_test, dt.predict_proba(X_test)[:, 1], 'Árvore de Decisão')
plot_roc_curve(y_test, svm.decision_function(X_test), 'SVM')
plot_roc_curve(y_test, nn.predict_proba(X_test)[:, 1], 'Redes Neurais')

# Ajuste do layout
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Aleatório')
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC para Modelos de Detecção de Fraude')
plt.legend()
plt.show()
```

```python
# DataFrame para armazenar as métricas
metricas_df = pd.DataFrame({
    'Modelo': modelos,
    'AUC-ROC': roc_auc,
    'Precisão': precision,
    'Recall': recall,
    'F1-Score': f1
})

# Exiba o DataFrame de métricas
print(metricas_df)

# Determine o melhor modelo com base em uma métrica específica
melhor_modelo_auc = metricas_df.loc[metricas_df['AUC-ROC'].idxmax()]
melhor_modelo_precisao = metricas_df.loc[metricas_df['Precisão'].idxmax()]
melhor_modelo_recall = metricas_df.loc[metricas_df['Recall'].idxmax()]
melhor_modelo_f1 = metricas_df.loc[metricas_df['F1-Score'].idxmax()]

print("\nMelhor Modelo (AUC-ROC):")
print(melhor_modelo_auc)

print("\nMelhor Modelo (Precisão):")
print(melhor_modelo_precisao)

print("\nMelhor Modelo (Recall):")
print(melhor_modelo_recall)

print("\nMelhor Modelo (F1-Score):")
print(melhor_modelo_f1)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=metricas_df, x='Modelo', y='AUC-ROC', palette='viridis')
plt.title('AUC-ROC para Modelos de Detecção de Fraude')
plt.show()
```

```python
# Crie um gráfico de barras para as métricas de precisão, recall e F1-Score
plt.figure(figsize=(12, 8))
metricas_df_melted = pd.melt(metricas_df, id_vars='Modelo', var_name='Métrica', value_name='Valor')
sns.barplot(data=metricas_df_melted, x='Modelo', y='Valor', hue='Métrica', palette='viridis')
plt.title('Comparação de Métricas para Modelos de Detecção de Fraude')
plt.ylabel('Valor')
plt.show()
```
